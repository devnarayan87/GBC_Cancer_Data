```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

# --- 1. Data Input (The text you provided) ---
# NOTE: This string contains the 10 tables of results.
# In a real script, this data would come directly from the cv_results dictionary.
data_text = """
Tuned model evaluation results:
                     Accuracy  Precision    Recall  F1 Score       AUC  \
Logistic Regression  0.858696   0.846154  0.897959  0.871287  0.892620
Random Forest        0.940217   0.957895  0.928571  0.943005  0.987067
SVM                  0.864130   0.841121  0.918367  0.878049  0.881941
KNN                  0.875000   0.831858  0.959184  0.890995  0.923410
Decision Tree        0.891304   0.890000  0.908163  0.898990  0.890128
AdaBoost             0.945652   0.940000  0.959184  0.949495  0.984931
CatBoost             0.934783   0.930000  0.948980  0.939394  0.989321
XGBoost              0.940217   0.939394  0.948980  0.944162  0.985643
LightGBM             0.934783   0.947917  0.928571  0.938144  0.989084
GradientBoosting     0.934783   0.947917  0.928571  0.938144  0.982321
Stacking             0.945652   0.968085  0.928571  0.947917  0.993830

Tuned model evaluation results:
                     Accuracy  Precision  Recall  F1 Score       AUC  \
Logistic Regression  0.891304   0.908163    0.89  0.898990  0.946071
Random Forest        0.956522   0.979167    0.94  0.959184  0.990774
SVM                  0.913043   0.937500    0.90  0.918367  0.949524
KNN                  0.875000   0.829060    0.97  0.894009  0.909167
Decision Tree        0.896739   0.909091    0.90  0.904523  0.910000
AdaBoost             0.967391   0.960784    0.98  0.970297  0.994821
CatBoost             0.961957   0.989474    0.94  0.964103  0.992500
XGBoost              0.956522   0.989362    0.93  0.958763  0.990952
LightGBM             0.967391   0.979592    0.96  0.969697  0.996071
GradientBoosting     0.961957   0.979381    0.95  0.964467  0.993333
Stacking             0.956522   0.979167    0.94  0.959184  0.994762

Tuned model evaluation results:
                     Accuracy  Precision  Recall  F1 Score       AUC  \
Logistic Regression  0.891304   0.908163    0.89  0.898990  0.946071
Random Forest        0.956522   0.979167    0.94  0.959184  0.990774
SVM                  0.913043   0.937500    0.90  0.918367  0.949524
KNN                  0.875000   0.829060    0.97  0.894009  0.909167
Decision Tree        0.896739   0.909091    0.90  0.904523  0.910000
AdaBoost             0.967391   0.960784    0.98  0.970297  0.994821
CatBoost             0.961957   0.989474    0.94  0.964103  0.992500
XGBoost              0.956522   0.989362    0.93  0.958763  0.990952
LightGBM             0.967391   0.979592    0.96  0.969697  0.996071
GradientBoosting     0.961957   0.979381    0.95  0.964467  0.993333
Stacking             0.956522   0.979167    0.94  0.959184  0.994762

Tuned model evaluation results:
                     Accuracy  Precision    Recall  F1 Score       AUC  \
Logistic Regression  0.885870   0.841121  0.957447  0.895522  0.928487
Random Forest        0.961957   0.967742  0.957447  0.962567  0.988416
SVM                  0.885870   0.828829  0.978723  0.897561  0.920449
KNN                  0.853261   0.791304  0.968085  0.870813  0.922636
Decision Tree        0.934783   0.927083  0.946809  0.936842  0.942435
AdaBoost             0.907609   0.888889  0.936170  0.911917  0.968440
CatBoost             0.956522   0.947917  0.968085  0.957895  0.989598
XGBoost              0.951087   0.956989  0.946809  0.951872  0.989480
LightGBM             0.945652   0.946809  0.946809  0.946809  0.989953
GradientBoosting     0.951087   0.956989  0.946809  0.951872  0.987116
Stacking             0.945652   0.946809  0.946809  0.946809  0.993144

Tuned model evaluation results:
                     Accuracy  Precision    Recall  F1 Score       AUC  \
Logistic Regression  0.836957   0.775701  0.932584  0.846939  0.878179
Random Forest        0.913043   0.876289  0.955056  0.913978  0.974571
SVM                  0.853261   0.798077  0.932584  0.860104  0.871674
KNN                  0.853261   0.787037  0.955056  0.862944  0.901478
Decision Tree        0.907609   0.867347  0.955056  0.909091  0.928681
AdaBoost             0.902174   0.873684  0.932584  0.902174  0.964636
CatBoost             0.918478   0.885417  0.955056  0.918919  0.976345
XGBoost              0.902174   0.865979  0.943820  0.903226  0.969486
LightGBM             0.923913   0.894737  0.955056  0.923913  0.971851
GradientBoosting     0.918478   0.885417  0.955056  0.918919  0.972206
Stacking             0.945652   0.934066  0.955056  0.944444  0.981195
"""

# --- 2. Data Parsing Function ---
def parse_cv_results_to_df(data_text):
    """Parses the multiple result tables into a single long-format DataFrame."""
    all_dfs = []

    # Split the text by the header line 'Model evaluation results:'
    sections = data_text.strip().split('Tuned model evaluation results:')

    # Define the expected columns (metrics)
    metric_cols = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']

    # Skip the first empty section
    for i, section in enumerate(sections[1:]):
        if not section.strip():
            continue

        # Read the table using pandas read_csv with StringIO.
        try:
            df = pd.read_csv(
                io.StringIO(section.strip()),
                # Use raw string r'' for regex and multiple spaces as separator
                sep=r'\s\s+',
                engine='python',        # Required for multi-character separators
                skiprows=1,             # Skip the misaligned header row
                header=None,            # Tell pandas there is no header
                skipinitialspace=True   # Skip spaces after delimiter
            )
        except Exception as e:
            print(f"Error parsing section {i}: {e}")
            continue

        # Check if the DataFrame has the expected number of columns (Model name + 5 metrics)
        if df.shape[1] < len(metric_cols) + 1:
            # Handle cases where the first column (model name) is missing/misaligned
            if df.shape[1] == len(metric_cols):
                 # If model names got merged into the index, we treat the index as the model names
                 df.index.name = 'Model'
                 df = df.reset_index()
            else:
                 print(f"Skipping section {i} due to unexpected column count ({df.shape[1]}).")
                 continue

        # Rename columns: the first column is the Model name, the rest are metrics
        if df.columns[0] != 'Model':
            df.columns = ['Model'] + metric_cols

        # Set 'Model' as the index for concatenation
        df = df.set_index('Model')

        all_dfs.append(df)

    # Concatenate all fold results and melt into long format
    full_df = pd.concat(all_dfs, keys=range(len(all_dfs)), names=['Fold', 'Model'])

    # Reset index and melt
    full_df = full_df.reset_index()
    long_df = pd.melt(full_df, id_vars=['Fold', 'Model'],
                      value_vars=metric_cols,
                      var_name='Metric', value_name='Score')

    # Convert 'Score' to numeric, forcing errors to NaN
    long_df['Score'] = pd.to_numeric(long_df['Score'], errors='coerce')

    # Drop any rows that resulted in NaN scores due to parsing issues
    long_df.dropna(subset=['Score'], inplace=True)

    return long_df

# --- 3. Execute Parsing and Plotting ---
results_long_df = parse_cv_results_to_df(data_text)

if results_long_df.empty:
    print("Could not parse any valid data. Please check the input format.")
else:
    # --- Aesthetics Settings ---
    sns.set_theme(style="whitegrid", rc={"axes.edgecolor": "#eaeaea"})

    # Create a figure with 5 subplots (one for each metric)
    metrics = results_long_df['Metric'].unique()
    n_metrics = len(metrics) # 5 metrics

    # NEW: Use 2 rows and 3 columns for better paper layout. Increased height for spacing.
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10), sharey=True)

    # Flatten the 2x3 axes array for easier indexing (axes[0, 0], axes[0, 1], ...)
    flat_axes = axes.flatten()

    # fig.suptitle('Box Plot Comparison of Model Performance Across 10 Cross-Validation Folds',
    #              fontsize=18, fontweight='bold', y=0.98)

    # Generate a box plot for each metric
    for i, metric in enumerate(metrics):
        # Filter data for the current metric
        plot_data = results_long_df[results_long_df['Metric'] == metric]

        # Create the box plot using seaborn, referencing the flattened axes array
        sns.boxplot(ax=flat_axes[i],
                    x='Model',
                    y='Score',
                    data=plot_data,
                    palette="Pastel1",
                    saturation=0.9,
                    width=0.6,
                    boxprops={'alpha': 0.7, 'edgecolor': 'black', 'linewidth': 1.5},
                    medianprops={'color': 'red', 'linewidth': 2},
                    whiskerprops={'color': 'black', 'linewidth': 1.5},
                    capprops={'color': 'black', 'linewidth': 1.5}
                    )

        # Add titles and labels
        flat_axes[i].set_title(f'{metric}', fontsize=15, fontweight='semibold')
        #flat_axes[i].set_xlabel("Model", fontsize=13)

        # Rotate x-labels to prevent overlap (65 degrees)
        for tick in flat_axes[i].get_xticklabels():
            tick.set_rotation(65)
            tick.set_fontsize(10)

        # Set y-axis label only for the first plot in each row (i=0 and i=3)
        # 0 is the first plot in row 1, 3 is the first plot in row 2
        if i == 0 or i == 3:
            flat_axes[i].set_ylabel("Score Value", fontsize=13, fontweight='medium')
        else:
            flat_axes[i].set_ylabel("") # Remove redundant y-label

        flat_axes[i].grid(axis='y', linestyle='--', alpha=0.5)

    # Hide the unused subplot (the 6th spot, index 5, in the 2x3 grid)
    if n_metrics < len(flat_axes):
        for j in range(n_metrics, len(flat_axes)):
            fig.delaxes(flat_axes[j])

    # Adjust layout to accommodate two rows and rotated labels (increased hspace and bottom margin)
    plt.subplots_adjust(top=0.90, bottom=0.15, hspace=0.8, wspace=0.1)

    # Save the figure as a high-quality PDF (vector graphics)
    # Use bbox_inches='tight' to ensure rotated labels are not clipped
    plt.savefig('model_comparison_boxplots_compact.pdf', format='pdf', dpi=300, bbox_inches='tight')

    # Save the figure as a PNG for display/web use
    plt.savefig('model_comparison_boxplots_compact.png', dpi=300)

    # Display the figure
    plt.show()

    print("The compact box plot visualization has been saved as both 'model_comparison_boxplots_compact.png' (for screen viewing) and 'model_comparison_boxplots_compact.pdf' (for high-quality, print-ready output).")
```
